```python
class DJB2:
    def __init__(self, initVal=5381):
        self._state = initVal
        
    def hash(self, val):
        x = (self._state << 5) & 0xFFFFFFFF
        x = (x + self._state) & 0xFFFFFFFF
        self._state = x ^ (val & 0xFFFFFFFF)
        return self._state
```
